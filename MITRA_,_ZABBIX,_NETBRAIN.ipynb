# ZABBIX data generation


```python
# create new provider class

class sevProvider(BaseProvider):
    def sev(self):
        return random.choice(['Warning', 'High', 'Average'])
#then add new provider to faker instance
fake.add_provider(sevProvider)
# now you can use:
sev = fake.sev()

class statusProvider(BaseProvider):
    def status(self):
        return random.choice(['RESOLVED'])
# then add new provider to faker instance
fake.add_provider(statusProvider)
# now you can use:
status = fake.status()
# create new provider class

class problemProvider(BaseProvider):
    def problem(self):
        return random.choice(['Link is down','System status is in warning/critical state', 'Device temperature is too high / too low', 'Power supply is in critical state', 'Free disk space is low',
       'Fan is in critical state', 'No SNMP data collection'])
# then add new provider to faker instance
fake.add_provider(problemProvider)
# now you can use:
problem = fake.problem()

class ackProvider(BaseProvider):
    def ack(self):
        return random.choice(['No'])
# then add new provider to faker instance
fake.add_provider(ackProvider)
# now you can use:
ack = fake.ack()

class actionsProvider(BaseProvider):
    def actions(self):
        return random.choice(['nan'])
# then add new provider to faker instance
fake.add_provider(actionsProvider)
# now you can use:
actions = fake.actions()

class tagsProvider(BaseProvider):
    def  tags(self):
        return random.choice(['Type: ASA, Vendor: Cisco', 'nan'])
# then add new provider to faker instance
fake.add_provider( tagsProvider)
# now you can use:
tags = fake. tags()
```

```python
def generate_bleu():
    t= fake.date_time_this_month(before_now=True, after_now=False,tzinfo=None)
    duration= datetime.timedelta(hours=random.randint(0, 18),minutes=random.randint(0, 60),seconds=random.randint(0, 60))
    rt= t + duration
    return np.concatenate([
        [
            fake.sev(),t,rt,duration,fake.status(),
            fake.pystr_format(string_format='??????????##', letters='JKLMNOPQRSTWXYZ'), fake.problem(),fake.ack(),fake.actions(),fake.tags()
        ]
        ])

with open('zabbix_fake.csv', 'w') as csvfile:
    writer = csv.writer(csvfile)
    writer.writerow(['Severity','Time', 'Recovery Time', 'Duration','Status', 'Host','Problem','Ack','Actions','Tags'])
    for n in range(1, 10000):
        writer.writerow(generate_bleu())
```

```python
#DF devices 1 
fwl1=net_fake.loc[net_fake.Dev1_type=="FWL"][["Device1"]]
fwl1.rename(columns={"Device1":"FWL"},inplace=True)
#DF devices 2 
fwl2=net_fake.loc[net_fake.Dev2_type=="FWL"][["Device2"]]
fwl2.rename(columns={"Device2":"FWL"},inplace=True)
#Concaténation de la liste des FWL 
fwl=pd.concat([fwl1, fwl2])
fwl
```

```python
#Séléction des IP et pays des devices hors que FWL du dataset mitra 
ip_country=mitra_fake.loc[mitra_fake.Techno!="FWL"][["IP","Country"]]
```

```python
class nameProvider(BaseProvider):
    def name(self):
        name=random.choice(fwl["FWL"].to_list())
        return name
# then add new provider to faker instance
fake.add_provider(nameProvider)
devname = fake.name()

class srcportProvider(BaseProvider):
    def srcport(self):
        srcport=random.choice(range(20000, 60000))
        return srcport
# then add new provider to faker instance
fake.add_provider(srcportProvider)
# now you can use:
srcport = fake.srcport()

class dstportProvider(BaseProvider):
    def dstport(self):
        dstport=random.choice(range(300, 10000))
        return dstport
# then add new provider to faker instance
fake.add_provider(dstportProvider)
# now you can use:
dstport = fake.dstport()

class typeProvider(BaseProvider):
#The weight of traffic type is the highest
    def type(self):
        return random.choice(random.choices(['traffic', 'utm', 'dns'], weights=(70, 25, 5)))
#then add new provider to faker instance
fake.add_provider(typeProvider)
# now you can use:
type_ = fake.type()

class ip_countrProvider(BaseProvider):

    def ip_countr(self):
        srcip= random.choice([i for i in ip_country["IP"]])
        dstip=random.choice([i for i in ip_country["IP"] if i not in [srcip]])
        return (srcip,dstip)

# then add new provider to faker instance
fake.add_provider(ip_countrProvider)
# now you can use:
ip_countr = fake.ip_countr()

class actionsdurationProvider(BaseProvider):
    def action(self):
        action=random.choice(['accept', 'deny', 'pass','close','client-rst','dns','ip-conn','pass','server-rst','timeout'])
        return action
# then add new provider to faker instance
fake.add_provider(actionsdurationProvider)
# now you can use:
action = fake.action()

class crlevelProvider(BaseProvider):
    def crlevel(self):
        return random.choice(random.choices(['high', 'low']))
#then add new provider to faker instance
fake.add_provider(crlevelProvider)
# now you can use:
crlevel= fake.crlevel()
```

```python
def action_related(row):
    #print (type(row))
    #print (row)
    if row.action=='ip-conn'or row.action=='pass' or row.action== 'dns':
            srccountry=''
            dstcountry=''
            duration=''
            sentbyte=''
            rcvdbyte=''
            sentpkt=''
    elif row.action=='deny':
            #print (row.srcip)
            srccountry= ip_country.loc[ip_country.IP==row.srcip]["Country"].values[0]
            dstcountry= ip_country.loc[ip_country.IP==row.dstip]["Country"].values[0]
            duration=0.0
            sentbyte=0.0
            rcvdbyte=0.0
            sentpkt=0.0
    else:
            #print (row.srcip)
            srccountry= ip_country.loc[ip_country.IP==row.srcip]["Country"].values[0]
            dstcountry= ip_country.loc[ip_country.IP==row.dstip]["Country"].values[0]   
            duration=random.choice(range(100, 10000))
            sentbyte=random.choice(range(100, 10000))
            rcvdbyte=random.choice(range(100, 10000))
            sentpkt=random.choice(range(100, 10000))
    row.srccountry = srccountry
    row.dstcountry= dstcountry
    row.duration= duration
    row.sentbyte= sentbyte
    row.rcvdbyte= rcvdbyte
    row.sentpkt= sentpkt
    return row
```

```python
def generate_syslog():
    t= fake.date_time_this_month(before_now=True, after_now=False,tzinfo=None)
    service=fake.pystr_format(string_format='service###')
    return np.concatenate([
        [
            t,fake.name(),fake.type(),fake.srcport(),fake.dstport(),service,fake.action(),fake.crlevel()
        ],fake.ip_countr()
        ])

def create_fake_df():
    l=[]
    for n in range(0, 1000):
        l.append(generate_syslog())
    return pd.DataFrame(l,columns=['time','devname', 'type','srcport','dstport','service','action','crlevel','srcip','dstip'])

        
```

```python
syslog_fake = create_fake_df()
syslog_fake['srccountry'] = ""
syslog_fake['dstcountry'] = ""
syslog_fake['duration'] = ""
syslog_fake['sentbyte'] = ""
syslog_fake['rcvdbyte'] = ""
syslog_fake['sentpkt'] = ""
```
